```python
dfDate['Date'] = pd.to_datetime(dfDate['Date'], format = '%d-%m-%Y')
dfDate
```

```python
dfDate['Trimestre'] = pd.PeriodIndex(dfDate['Date'], freq = 'Q')
dfDate
```

```python
dfDate.head(50)
```

```python
venTrim = dfDate.groupby(['Store', 'Trimestre'])['Weekly_Sales'].mean().reset_index()
venTrim.head(20)
```

```python
ventasPivot = venTrim.pivot(index = 'Store', columns = 'Trimestre', values = 'Weekly_Sales')
ventasPivot.head(20)
```

```python
trimAnt = ventasPivot.columns[-3]
trimAnt
#trimAct = ventasPivot.columns[-1]
#tasaCre = ventasPivot[trimAct].pct_change(periods = 1)/ventasPivot[trimAnt]
#tasaCre
```

```python
trimAct = ventasPivot.columns[-2]
trimAct
```

```python
#tasaCre = ventasPivot[trimAct].pct_change(periods = 1)/ventasPivot[trimAnt]
tasaCre = (ventasPivot[trimAct] - ventasPivot[trimAnt]) / ventasPivot[trimAnt] * 100
tasaCre
```
